```python
file_path = mock_csv_data(data, tmp_path="data", file_name="corral")
```

```python
data_prep = DataPreparation(
    file_path=file_path, x_cols=list(data.columns.difference(["target"])), y_col="target"
)
```

```python
data_prep
```

```python
x, y = data_prep.select_columns()
x_train, x_test, y_train, y_test = data_prep.prepare_data()
data_dict = {
    "train_x": x_train,
    "test_x": x_test,
    "train_y": y_train,
    "test_y": y_test,
}
```

```python
from quoptuna import Optimizer
```

```python
file_name = "corral"
```

```python
optimizer = Optimizer(db_name=file_name, study_name=file_name, data=data_dict)
```

```python
study, best_trials = optimizer.optimize(n_trials=100)
```
